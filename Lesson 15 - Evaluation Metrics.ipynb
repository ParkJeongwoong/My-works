## Precision & Recall
> 위의 정확도의 단점을 보완하기 위한 두 지표

### Precision

정밀도 : 긍정(Positive)적 예측 중 실제 정답으로 맞출 확률\
True Positive / (True Positive + False Positive)

### Recall

재현율 : 실제 데이터 속 정답을 의미하는 예측 중 실제로 정답을 맞출 확률\
True Positive / (True Positive + Flase Negative)
