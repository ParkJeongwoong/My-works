## Bag of Words (aka CounterVectorizer)
'Nice', 'Day'\
'A', 'Very'\
이런 글자들이 Dictionary 형태로 모인 Bag => `Value`는 단어 등장 빈도수

- 단어의 순서는 상관 X
- 단어의 길이는 상관 O / Input Vector에 영향
- 복잡한 경우의 수 구분 X

# Weighting by Tern Frequency
## Tf-Idf Representation
Tf : Term Frequency\
Idf : Inverse Document Frequency

Tf <- sth like bag of words\
Idf <- `weighting by how often word occurs in corpus`(말뭉치)

`Tf-IdF에서는 Rare Word를 Common Word 보다 더 높은 가중치를 준다`
