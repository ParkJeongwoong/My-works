# Outlier

> linear regression 등의 예측 선에서 벗어난 점

# Outlier의 원인
- Sensor malfunction -> 무시
- Data entry errors -> 무시
- <u>Freak event</u> -> **여기에 집중해야 함**

# Outlier의 종류

4번째는 약간 논란의 여지가 있음 / 남자 분은 Outlier라고, 여자 분은 아니라고 함 (근데 영상이 남자 자기 영상이라 답에 포함시켰다고 함)

# Outlier Detection&Removal*****(핵심)

1. Train
2. `Remove 10% of data points which have highest error`
3. `Train Again`
4. Repeat 2 and 3

1번을 진행하면 위와 같은 Regression line이 만들어 진다

10%를 제거하는 과정에서 맨 아래에 있던 한 개의 점이 지워지게 되고 위와 같은 선으로 바뀐다

## Residual

잔차 VS 오차

- 잔차(Residual): 표본에서의 차이값
- 오차(Error): 모집단에서의 차이값
