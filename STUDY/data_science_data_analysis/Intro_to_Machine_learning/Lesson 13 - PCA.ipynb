# Principal Component Analysis

- `고차원의 데이터를 저차원의 데이터로 환원시키는 기법`

> 거의 모든 종류의 데이터 분석에 삭제됨 (Set Compression을 포함해서)


이런 경우도 1 차원 그래프라고 할 수 있음


이렇게 remapping이 가능하기 때문

- PCA 에서는 축을 shift, rotate만 가능 (translation & rotation only)
- PCA는 새로운 축을 만든다
    1. 중심축을 Data의 중심으로 이동 => `Center 좌표 찾기`
    2. variation의 수직 방향으로 X축을 이동 => `x'의 Vector 찾기`
    3. X축과 수직으로 Y축 형성 => `y'의 Vector 찾기`


이렇게 만들면 x', y' Vector가 형성된다

## Data ready for PCA
> 모든 데이터 셋에서 Center를 찾고 x', y' 벡터를 찾는 것은 가능하다

**하지만!!!**, 제대로 활용을 하려면 <u>major axix (x')</u>가 dominate한 상태여야 한다


## Measurable vs Latent Features


정답: `Linear Regression`

이유: output(`price`)이 continuous 하기 때문 => classifier를 쓰기 좋지 않다

- Latent한 정보를 얻기 위해 Measurable한 정보를 참조해야 한다
- 문제를 간단히 만들기 위해 이런 measuable 정보들을 압축해서 몇 가지만 남겨 사용

정답: `Linear Regression`

이유: 정확히 몇 개의 특징을 사용할지 정할 수 있기 때문 (K를 2로 정하면 가장 효과적인 특징 2개만 남기고 다 버림)

SelectPercentile은 정확히 몇 개의 feature가 있는지 모르기 때문에 좋은 선택이 아니다.\
(ex. feature가 4개라면 4개인 걸 알아야 50%를 선택해서 2개를 남길지 선택할 수 있다)

=> 근데 그냥 이렇게 feature를 줄이는 것 보다 `composite feature`를 만드는 것이 낫다\
이게 바로 `Principal Component`

방법은 아까 말했듯, <u>차원을 낮추는 것</u>


이렇게 만든 principal component 위로 data를 projection 한다.
