# Feature Selection
> 문제를 간단하게 만들기
1. select best features
2. add new features

## Feature 제거
[이유]
- overfitting 발생 가능
- 이미 존재하는 feature와 동일한 영향을 끼치는 feature는 필요 없음
- feature가 많아질 수록 training/testing 속도가 느려짐

즉 feature는 information이 아니다 (정보를 얻기 위한 원석일 뿐)

## Univariate Feature Selection
statistical test를 통해 feature간의 관계 이해 가능\
sklearn의 `SelectPercentile`과 `SelectKBest`를 통해 feature selection 가능

## Bias, Variance, # of Features

`high bias` : data에 너무 적은 관심을 쏟는 것 => oversimplified | **training set에 대해 error가 높다** `적은 feature 사용`\
`high variance` : data에 너무 많은 관심을 쏟는 것 => overfits | **testing set에 대해 error가 높다** `많은 feature 사용`
