# Feature Scaling
- 몇몇 특징을 뽑아 계산하는 방법 (ex. Size = height + weight)

이 때 중요한 것은, 연관식을 잘 짜야 한다는 것! (계수, 지수 등을 최적화 시켜야 한다)

따라서 `Rescaling`이 필요!

## Feature Scaling Formula

x' = (x-x_min) / (x_max-x_min)

이렇게 해 주면 항상 0~1 사이의 값이 나오게 된다

- 장점 : 항상 일정한 값이 나온다
- 단점 : Outlier의 존재로 인해 정확도가 떨어질 수 있다
